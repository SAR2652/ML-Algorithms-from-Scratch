```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from Tensorflow.Linear_Models.logistic_regression_tf import LogisticRegression
```

#### 10,000 samples and 4 features

```python
m = 10000
n = 4
```

```python
X, y = make_classification(m, n)
```

```python
X.shape
```

```python
y.T.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.225, random_state = 42)
```

```python
lr = LogisticRegression(verbose = True)
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_val, y_val)
```

```python
lr.plot_cost_values()
```

```python
lr.plot_weights()
```

```python
lr.plot_bias()
```

```python
lr.plot_accuracy()
```
